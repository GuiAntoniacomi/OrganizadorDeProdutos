```python
import pandas as pd

arquivo = pd.read_excel(r"C:\Users\anton\OneDrive\Área de Trabalho\Documentos e Planilhas\Power Bi\Geral Loja\2024\Vendas 2024.xlsx")
df_vendas = pd.DataFrame(arquivo)
```

```python
json = pd.read_json(r'Base\bagy.json')
df_bagy = pd.DataFrame(json)
json.head()
```

```python
df_bagy = df_bagy[['Brands → Name', 'Categories - Category Default → Name', 'Price', 'Cost', 'Slug', 'Name', 'Stocks → Balance', 'External ID', 'Variations → Sku']]
df_bagy = df_bagy.rename(columns={'Brands → Name': 'Marca', 'Categories - Category Default → Name': 'Categoria', 'Price': 'Preço', 'Cost': 'Custo', 'Slug': 'Link', 'Name':'Produto', 'Stocks → Balance': 'Estoque', 'External ID': 'SKU Pai', 'Variations → Sku': 'Código'})
ordem_colunas = ['SKU Pai', 'Código', 'Produto', 'Link', 'Estoque', 'Marca', 'Categoria', 'Preço', 'Custo']
df_bagy = df_bagy[ordem_colunas]
df_bagy = df_bagy.dropna(subset='Código')
df_bagy['Código'] = df_bagy['Código'].astype(int)
df_bagy
```

```python
df_bagy = df_bagy[['SKU Pai', 'Código']]
df_bagy
```

```python
df_vendas = df_vendas.merge(df_bagy, how='left', on='Código')
ordem_colunas = ['SKU Pai', 'Código', 'Nome da Loja', ' Valor total ', 'Custo do Produto', 'Cidade', 'UF']
df_vendas = df_vendas[ordem_colunas]
```

```python
df_vendas[' Valor total '] = df_vendas[' Valor total '].replace(',', '.')
df_vendas['Custo do Produto'] = df_vendas['Custo do Produto'].replace(',', '.')
```

```python
df_vendas[' Valor total '] = df_vendas[' Valor total '].astype(float)
df_vendas['Custo do Produto'] = df_vendas['Custo do Produto'].astype(float)
df_vendas
```

```python
df_vendas['Nº De Vendas'] = 1
df_vendas = df_vendas.groupby("Código").agg({
    'SKU Pai': 'first',
    'Nome da Loja': 'first',
    ' Valor total ': 'sum',
    'Custo do Produto': 'sum',
    'Cidade': 'first',
    'UF': 'first',
    'Nº De Vendas':'sum'
}).reset_index()
df_vendas
```

```python
df_vendas = df_vendas.sort_values(by='Nº De Vendas', ascending=False)
df_vendas
```

```python
df_vendas = df_vendas[['Código', 'Nº De Vendas']]
df_vendas
```
