```python
import pandas as pd

json = r'C:\Users\anton\Downloads\todos_os_produtos_2024-06-04T13_19_23.142743Z.json'

df = pd.read_json(json)
df_bagy = df[['Brands → Name', 'Categories - Category Default → Name', 'Price', 'Cost', 'Slug', 'Name', 'Stocks → Balance', 'External ID']]
df_bagy
```

```python
df_bagy = df_bagy.rename(columns={'Brands → Name': 'Marca', 'Categories - Category Default → Name': 'Categoria', 'Price': 'Preço', 'Cost': 'Custo', 'Slug': 'Link', 'Name': 'Produto', 'Stocks → Balance': 'Estoque', 'External ID': 'SKU Pai'})
df_bagy['Total Preço'] = df_bagy['Preço'] * df_bagy['Estoque']
df_bagy['Total Custo'] = df_bagy['Estoque'] * df_bagy['Custo']
df_bagy['MKP'] = df_bagy['Preço'] / df_bagy['Custo']
df_bagy['MKP'] = df_bagy['MKP'].round(2)
df_bagy['Marca'].dropna()
ordem_colunas = ['SKU Pai', 'Produto', 'Link', 'Estoque', 'Marca', 'Categoria', 'Preço', 'Custo', 'MKP', 'Total Preço', 'Total Custo']
df_bagy = df_bagy[ordem_colunas]
df_bagy
```

```python
mapeamento_categorias = {
    'camiseta': 'Camisetas',
    'Boné': 'Bonés',
    'Polos': 'Polos',
    'Calças': 'Calças',
    'Bermuda': 'Bermudas',
    'Tênis': 'Tênis',
    'Blusas': 'Moda Inverno',
    'Camisa': 'Camisas Sociais',
    'Carteira': 'Carteiras',
    'Cueca': 'Cuecas',
    'Chinelo': 'Chinelos',
    'Cinto': 'Cintos',
    'Meia': 'Meias',
    'Pijama': 'Pijamas',
    'Mochila': 'Mochilas e Malas',
    'Sunga': 'Sungas',
    'Calçado': 'Calçados',
    'Aramis': 'Acessórios',
    'King & Joe': 'Chapéus',
    'Marcas': 'Acessórios',
    'Polo Ralph Lauren': 'Polos',
    'U.S. Polo Assn.': 'Polos',
    'Reserva': 'Acessórios',
    'Sergio K': 'Acessórios'
}
for termo, nova_categoria in mapeamento_categorias.items():
    df_bagy.loc[df_bagy['Categoria'].str.contains(termo, case=False, na=False), 'Categoria'] = nova_categoria

df_bagy.dropna(subset= ['Categoria'], inplace=True)
df_bagy
```
