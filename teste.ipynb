```python
import pandas as pd

arquivo = pd.read_json(r"C:\Users\anton\Downloads\todos_os_produtos_2024-05-29T19_00_09.750649Z.json")

df_bagy = pd.DataFrame(arquivo)
```

```python
df_resumida = df_bagy[['Brands → Name', 'Categories - Category Default → Name', 'Price', 'Cost', 'Slug', 'Name', 'Stocks → Balance', 'External ID']]
df_resumida  = df_resumida.rename(columns={'Brands → Name': 'Marca', 'Categories - Category Default → Name': 'Categoria', 'Price': 'Preço', 'Cost': 'Custo', 'Slug': 'Link', 'Name': 'Produto', 'Stocks → Balance': 'Estoque', 'External ID': 'SKU Pai'})
ordem_colunas = ['SKU Pai', 'Produto', 'Link', 'Estoque', 'Marca', 'Categoria', 'Preço', 'Custo']
df_resumida = df_resumida[ordem_colunas]
df_resumida['Total Preço'] = df_resumida['Preço'] * df_resumida['Estoque'] 
df_resumida['Total Custo'] = df_resumida['Estoque'] * df_resumida['Custo']
df_resumida.head(5)
```

```python
df_agrupada = df_resumida.groupby('SKU Pai').agg({
    'Produto': 'first',
    'Link': 'first',
    'Estoque': 'sum',
    'Marca': 'first',
    'Categoria': 'first',
    'Preço': 'sum',
    'Custo': 'sum',
    'Total Preço': 'sum',
    'Total Custo': 'sum'
}).reset_index()
df_agrupada = df_agrupada.sort_values(by='Estoque', ascending=False).reset_index(drop=True)
df_agrupada['Marca'].dropna()

df_agrupada
```

```python
def remover_marca(categoria, marca):
    if pd.isna(categoria) or pd.isna(marca):
        return categoria
    # Substituir todas as ocorrências de 'marca' em 'categoria' e remover espaços extras
    return categoria.replace(marca, '').strip()

# Aplicar a função para remover a marca da coluna Categoria
df_agrupada['Categoria'] = df_agrupada.apply(lambda row: remover_marca(row['Categoria'], row['Marca']), axis=1)

# Exibir o DataFrame resultante
df_teste = df_agrupada[['Marca', 'Categoria']]
df_teste = df_teste.groupby('Categoria').agg({
    'Marca': 'first'
}).reset_index()

df_teste
```

```python
import pandas as pd

# Carregar o arquivo JSON
arquivo = pd.read_json(r"C:\Users\anton\Downloads\todos_os_produtos_2024-05-29T15_05_06.99827Z.json")

df_bagy = pd.DataFrame(arquivo)

df_resumida = df_bagy[['Brands → Name', 'Categories - Category Default → Name', 'Price', 'Cost', 'Slug', 'Name', 'Stocks → Balance', 'External ID']]
df_resumida = df_resumida.rename(columns={'Brands → Name': 'Marca', 'Categories - Category Default → Name': 'Categoria', 'Price': 'Preço', 'Cost': 'Custo', 'Slug': 'Link', 'Name': 'Produto', 'Stocks → Balance': 'Estoque', 'External ID': 'SKU Pai'})
ordem_colunas = ['SKU Pai', 'Produto', 'Link', 'Estoque', 'Marca', 'Categoria', 'Preço', 'Custo']
df_resumida = df_resumida[ordem_colunas]
df_resumida['Total Preço'] = df_resumida['Preço'] * df_resumida['Estoque']
df_resumida['Total Custo'] = df_resumida['Estoque'] * df_resumida['Custo']

df_agrupada = df_resumida.groupby('SKU Pai').agg({
    'Produto': 'first',
    'Link': 'first',
    'Estoque': 'sum',
    'Marca': 'first',
    'Categoria': 'first',
    'Preço': 'sum',
    'Custo': 'sum',
    'Total Preço': 'sum',
    'Total Custo': 'sum'
}).reset_index()
df_agrupada = df_agrupada.sort_values(by='Estoque', ascending=False).reset_index(drop=True)

# Lista de palavras-chave de categorias
categorias_possiveis = ['Camiseta', 'Bermuda', 'Cueca', 'Boné', 'Camisa', 'Calça', 'Jaqueta', 'Moletom', 'Blusa', 'Polo', 'Tênis', 'Sapato']

# Função para identificar a categoria com base nas palavras-chave
def identificar_categoria(categoria):
    if pd.isna(categoria):
        return categoria
    for palavra in categorias_possiveis:
        if palavra.lower() in categoria.lower():
            return palavra
    return categoria  # Retorna a categoria original se nenhuma palavra-chave for encontrada

# Aplicar a função para ajustar a coluna Categoria
df_agrupada['Categoria'] = df_agrupada['Categoria'].apply(lambda x: identificar_categoria(x))

# Exibir o DataFrame resultante
df_teste = df_agrupada[['Marca', 'Categoria']]
df_teste = df_teste.groupby('Categoria').agg({
    'Marca': 'first'
}).reset_index()
df_teste
```
