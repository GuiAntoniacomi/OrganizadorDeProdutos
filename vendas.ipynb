```python
import pandas as pd

arquivo = pd.read_excel(r"C:\Users\anton\OneDrive\Área de Trabalho\Documentos e Planilhas\Vendas Totais\Vendas Totais.xlsx")
df_vendas = pd.DataFrame(arquivo)
```

```python
display(df_vendas.head())
df_vendas.columns
```

```python
df_vendas = df_vendas[['Data', 'Nome da Loja', 'Marca', 'Código do produto pai',
       'Código (SKU)', 'Descrição do produto pai', 'Valor total', 'Preço de custo']]
```

```python
df_vendas = df_vendas.dropna(subset='Código (SKU)')
df_vendas['Código (SKU)'] = df_vendas['Código (SKU)'].astype(int)
df_vendas = df_vendas[df_vendas['Descrição do produto pai'] != '-']
df_vendas['MKP'] = df_vendas['Valor total'] / df_vendas['Preço de custo']
df_vendas
```

```python
df_vendas['Tipo de Produto'] = 'Coleção'
df_vendas.loc[df_vendas['MKP'] >= 2.6, 'Tipo de Produto'] = 'Saldo / Oportunidade'
df_vendas
```
